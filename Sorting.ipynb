```python
# Bubble sort 
list_demo = [12,45,65,92,11,32,4,46,76,98,90]
list_f = [12,5,65,9]
## we take range from 9 to 1 - fist outer ring will sort last value , one sort --puts one value in place . The last one will aotomatically be placed so we take one less than the total len 
# inner ring - we take 0 to outer ring -when outer range 3 it compares 3 time (0,1,2), when it is 2 it compares  2 times 
# after every outer loop the size of list is reduced by 1 

def bubble_sort(alist):
    for passum in range(len(alist)-1,0,-1):
        # print('outer range',passum)
        for i in range(passum):
            # print('inner range',i)
            # print('Compare-',i,alist[i],alist[i+1])
            if alist[i] > alist[i+1]:
                # print(f'compare',alist[i],alist[i+1])
                temp = alist[i]
                alist[i] = alist[i+1]
                alist[i+1] = temp
                # print(alist)
                # alist[i],alist[i+1] = alist[i+1],alist[i]
    return alist

bubble_sort(list_f)


# alist = [54,26,93,17,77,31,44,55,20]
# list_demo = [12,45,65,92,11,32,4,46,76,98]
# def bubble_Sort(alist):
#     for passum in range(len(alist)-1,0,-1):
#         # print('passum:',passum)
#         for item in range(passum):
#             # print('item:',item)
#             if alist[item] > alist[item+1]:
#                 # print('Compare:',alist[item],alist[item+1])
#                 alist[item],alist[item+1] = alist[item+1],alist[item]
#                 # print('List:',alist)
#     print('Final List:',alist)

# bubble_Sort(list_demo)

    
```

```python
demo = [12,45,65,92,11,32,4,46,76,98,90]
def bsort(alist):
    for passum in range(len(alist)-1,0,-1):
        # print('outer-loop:',passum)
        for i in range(passum):
            # print('inner-loop:',i)
            if alist[i] > alist[i+1]:
                alist[i],alist[i+1] = alist[i+1],alist[i]
    return alist
        

bsort(demo)
```

```python
# Insertion Sort 
demo = [12,5,65,9]
def insertion_sort(alist):
    for index in range(1,len(alist)):
        print('Outer Loop',index)
        current_val = alist[index]
        position = index
        print(f'Current Value: {current_val},Position: {position}')
        while position > 0 and alist[position-1] > current_val:
            alist[position] = alist[position-1]
            print(f"comparing: {alist[position] } ,{alist[position-1]}" )
            position = position-1
            print('new posotion',position)
            print('list:',alist)
        alist[position] = current_val

        return alist
    
        


insertion_sort(demo)
```

```python
demo = [12,5,65,9]
def insertion_sort(alist):
    for index_insertion in range(1,len(alist)):
        print('outer Loop:',index_insertion)
        current_value = alist[index_insertion]
        position = index_insertion
        print(f'current Val: {current_value},position: {position}')
        while position > 0 and alist[position-1] > current_value:
            print(f'compare:{alist[position-1]},{current_value}')
            alist[position] = alist[position-1]
            print(f'New List: {alist}')
            position = position-1
            print(f'New Position: {position}')
        alist[position] = current_value
    return alist
insertion_sort(demo)
```
